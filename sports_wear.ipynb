### 5-Data Preprocessing

```python
# Drop irrelevant columns
drop_cols = ['article', 'article.1', 'retailweek', 'customer_id']
data.drop(columns=drop_cols, inplace=True)
```

```python
data.head()
```

```python
# Handle sizes: Create binary columns for each size
size_set = ['xxs', 'xs', 's', 'm', 'l', 'xl', 'xxl']
for size in size_set:
    data[f'has_{size}'] = data['sizes'].apply(lambda x: int(size in str(x).lower().split(',')))
```

```python
data.columns
```

```python
data.head(10)
```

```python
# Drop the original sizes column
data.drop(columns=['sizes'], inplace=True)
```

```python
# categorical column
categorical=[]
for i in data.columns:
    if data[i].dtype=='object':
      categorical.append(i)
print(categorical)
```

```python
for i in categorical:
    print ('column name',i)
    print(data[i].unique())
    print('--'*40)
```

```python
# encoding data
from sklearn.preprocessing import LabelEncoder
Encoder=LabelEncoder()
for i in categorical:
    data[i]=Encoder.fit_transform(data[i])
```

```python
data.head()
```

```python
#  split data into feature(x) and target(y)
x = data.drop('label', axis=1)
y = data['label']
```

```python
# split data into data_train and data_test
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
print('x_train shape :',x_train.shape)
print('y_train shape :',y_train.shape)
```

```python
print('x_test shape :',x_test.shape)
print('y_test shape :',y_test.shape)
```

```python
data['label'].value_counts()
```

```python
# Rebalance train data
from imblearn.over_sampling import SMOTE

s = SMOTE(random_state=42)
x_train, y_train = s.fit_resample(x_train, y_train)
```

### 6-Aplay Model

```python
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, ExtraTreesClassifier, BaggingClassifier, AdaBoostClassifier, GradientBoostingClassifier, VotingClassifier
from xgboost import XGBClassifier
from sklearn.model_selection import GridSearchCV, cross_val_score, StratifiedKFold
```

```python
# Single Models
Single_models = {
    'Logistic Regression': LogisticRegression(),
    'KNN': KNeighborsClassifier(),
    'Naive Bayes': GaussianNB(),
    'SVM': SVC(),
    'Decision Tree': DecisionTreeClassifier()
}

# Ensemble Models
ensemble_models = {
    'Random Forest': RandomForestClassifier(),
    'Extra Trees': ExtraTreesClassifier(),
    'Bagging': BaggingClassifier(),
    'AdaBoost': AdaBoostClassifier(),
    'Gradient Boosting': GradientBoostingClassifier(),
    'XGBoost': XGBClassifier()
}

# Hyperparameter Grids
param_grids = {
    'Logistic Regression': {'C': [0.1, 1, 10]},
    'KNN': {'n_neighbors': [3, 5, 7]},
    'Naive Bayes': {}, # Add an empty dictionary for Naive Bayes since it doesn't usually require hyperparameter tuning
    'SVM': {'C': [0.1], 'kernel': ['linear']},
    'Decision Tree': {'max_depth': [None, 5, 10]},
    'Random Forest': {'n_estimators': [50, 100]},
    'Extra Trees': {'n_estimators': [50, 100]},
    'Bagging': {'n_estimators': [10, 20]},
    'AdaBoost': {'n_estimators': [50, 100]},
    'Gradient Boosting': {'n_estimators': [50, 100]},
    'XGBoost': {'n_estimators': [50], 'learning_rate': [0.01, 0.1]}
}
```

```python
# kfold
cv=StratifiedKFold(n_splits=5,shuffle=True,random_state=42)
```

```python
# Model Evaluation and Selection on single model
best_model={}
for name , model in Single_models.items():
    grid=GridSearchCV(model,param_grids[name],cv=cv,scoring='accuracy')
    grid.fit(x_train,y_train)
    best_model[name]=grid.best_estimator_
    print('model name = ',name)
    print('Best Paramter = ',grid.best_params_)
    print('Best score = ',grid.best_score_)
    print('--'*20)
```

```python
# Model Evaluation and Selection on enesmble model
for name,model in ensemble_models.items():
    grid=GridSearchCV(model,param_grids[name],cv=cv,scoring='accuracy',error_score='raise')
    grid.fit(x_train,y_train)
    best_model[name]=grid.best_estimator_
    print('model name = ',name)
    print('Best Paramter = ',grid.best_params_)
    print('Best score = ',grid.best_score_)
    print('--'*20)
```

```python
best_model
```

```python
# Evaluate Best Models on Test Set
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix

for name,model in best_model.items():
    y_pred=model.predict(x_test)
    cm=confusion_matrix(y_test,y_pred)
    print("model name =",name)
    print('accuracy score = ',accuracy_score(y_test,y_pred))
    print('classification_report = ',classification_report(y_test,y_pred))
    # plot confusion_matrix
    plt.figure(figsize=(5,5))
    sns.heatmap(cm,annot=True,fmt='.0f',cmap='Blues')
    plt.xlabel('Predicted')
    plt.ylabel('Actual')
    plt.title('Confusion Matrix')
    plt.show()
    print("--"*40)
```
